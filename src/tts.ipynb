```python
import edge_tts
import asyncio
import sounddevice as sd
import soundfile as sf
import os

TEXT = "Please adjust your volume until you can hear my voice clearly.  Click the next arrow when ready to continue!"

async def main():
    output_file = "volume_prompt.wav"

    # Generate TTS audio directly into the fixed file
    communicate = edge_tts.Communicate(TEXT, voice="en-US-AriaNeural")
    await communicate.save(output_file)

    # Load and play
    data, samplerate = sf.read(output_file)
    sd.play(data, samplerate)
    sd.wait()

await main()
```
